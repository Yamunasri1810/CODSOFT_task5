# Modelling

```python
train = new_data.drop("Class", axis=1)
test = new_data["Class"]
```

```python
train
```

```python
test
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, test, test_size=0.2,stratify=test, random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
X_train_pred = model.predict(X_train)
print(X_train_pred)
```

```python
training_accuract_score= accuracy_score(X_train_pred, y_train)
print(training_accuract_score)
```

```python
X_test_pred= model.predict(X_test)
print(X_test_pred)
```

```python
test_accuracy_score= accuracy_score(X_test_pred, y_test)
print(test_accuracy_score)
```
